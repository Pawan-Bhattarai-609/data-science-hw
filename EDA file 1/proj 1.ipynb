```python
sns.histplot(df['Age'],kde=False,
             bins=6,color='blue',
             )
#add labels
plt.title('Age count')
plt.xlabel('sex')
plt.ylabel('count')
```

-here we plot a histogram between count and sex on Y and X axis respectively

```python
sns.relplot(data=df, x="Age",y="Annual Income (k$)",kind="line")
```

-here we plot the line graph between age and annual income in X and Y axis respectively

```python
sns.scatterplot(x=df['Age'],y=df['Spending Score (1-100)'])
plt.title('spending score v age ')


```

-this is a scatter plot between age on X-axis and spending score on Y-axis

```python
import seaborn as sns
sns.barplot(x = 'Age',y='Gender',
            
        
            data = df)
```

-this is a bar graph between age on X-axis and gender on Y-axis

```python
sns.barplot(data=df, x="Gender",y="Spending Score (1-100)")
```

-this is a bar graph between gender on X-axis and spending score  on Y-axis

```python
sns.boxplot(data=df,x="Gender",y="Spending Score (1-100)")
```

-here we made box plot by taking gender as the parameter in X axis and spending score as the parameter in Y axis

```python

sns.scatterplot(x=df['Gender'],y=df['Annual Income (k$)'],hue=df['Age'])
```

# DATA SET ANALYSIS

- The histogram shows the distribution of ages in the dataset, with the count of individuals in different age bins.
- A line plot (sns.relplot) is used to visualize the relationship between 'Age' and 'Annual Income (k$)'.

 
- Bar plots are used to display the count of individuals for each age and gender combination.
- Another bar plot shows the average 'Spending Score (1-100)' for each gender.
- A box plot is created to compare the distribution of 'Spending Score (1-100)' between genders.
- Scatter plots are used to explore the relationship between 'Annual Income (k$)' and 'Age'. One scatter plot includes the 'Gender' dimension.
- A heatmap of the correlation matrix is plotted to visualize the correlations between numeric variables. 
- The data exploration provides insights into potential patterns and relationships between age, gender, income, and spending scores.
- Correlation analysis indicates whether there are significant linear relationships between numeric variables.
